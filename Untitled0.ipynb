```python
# ============================================
# SIMPLEST VERSION - Guaranteed to work
# ============================================

!pip install scikit-learn nltk pandas

import pandas as pd
import numpy as np
import re
import time
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('punkt_tab', quiet=True)

# Download dataset
print("Downloading dataset...")
!wget -q https://raw.githubusercontent.com/rasbt/python-machine-learning-book-3rd-edition/master/ch08/movie_data.csv.gz
!gunzip -f movie_data.csv.gz

df = pd.read_csv('movie_data.csv')
print(f"Dataset loaded: {len(df)} reviews")

# Use smaller subset for speed
train_texts = df['review'].tolist()[:5000]
train_labels = df['sentiment'].tolist()[:5000]
test_texts = df['review'].tolist()[5000:6000]
test_labels = df['sentiment'].tolist()[5000:6000]

print(f"Training: {len(train_texts)} reviews")
print(f"Testing: {len(test_texts)} reviews")

# Preprocessing
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess(text):
    text = text.lower()
    text = re.sub(r'<br\s*/?>', ' ', text)
    text = re.sub(r'[^a-z\s]', '', text)
    words = text.split()
    words = [lemmatizer.lemmatize(w) for w in words if w not in stop_words and len(w) > 2]
    return ' '.join(words)

print("Preprocessing...")
train_processed = [preprocess(t) for t in train_texts]
test_processed = [preprocess(t) for t in test_texts]

# TF-IDF
print("Creating features...")
vectorizer = TfidfVectorizer(max_features=3000)
X_train = vectorizer.fit_transform(train_processed)
X_test = vectorizer.transform(test_processed)

# Naive Bayes
print("\n--- Naive Bayes ---")
start = time.time()
nb = MultinomialNB()
nb.fit(X_train, train_labels)
nb_time = time.time() - start
nb_pred = nb.predict(X_test)
nb_acc = accuracy_score(test_labels, nb_pred)

print(f"Accuracy: {nb_acc:.3f} ({nb_acc*100:.1f}%)")
print(f"Precision: {precision_score(test_labels, nb_pred):.3f}")
print(f"Recall: {recall_score(test_labels, nb_pred):.3f}")
print(f"F1-Score: {f1_score(test_labels, nb_pred):.3f}")
print(f"Time: {nb_time:.1f} sec")

print("\n" + "="*50)
print("COPY THESE RESULTS:")
print(f"Naive Bayes Accuracy: {nb_acc*100:.1f}%")
print("="*50)
```
